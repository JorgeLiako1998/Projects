```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/yayoliak14/Desktop/ProFuturocsv.csv")
```

```python
df = df[df != "Null"].dropna()
```

```python
for col in df.columns[2:]:
    df[col] = df[col].str.rstrip('%').astype(float)
    df[col] = df[col] * 0.01
```

```python
df = df.melt(id_vars=['Plan', 'YEAR'], var_name='Month', value_name='Value')
```

```python
df.head()
```

```python
df['AVG'] = df.groupby(["Plan", "Month"])["Value"].transform('mean')
df['STDDEV'] = df.groupby(["Plan", "Month"])["Value"].transform('std')
df['MIN'] = df.groupby(["Plan", "Month"])["Value"].transform('min')
df['MAX'] = df.groupby(["Plan", "Month"])["Value"].transform('max')
df["DIFF"] = df["MAX"] - df['MIN']
```

```python
df.head(30)
```

```python
# Group by Month and get the index of the row with the highest average value for each month
highest_avg_idx = df.groupby('Month')['AVG'].idxmax()

# Use the index to retrieve the corresponding Plan
highest_avg_plan = df.loc[highest_avg_idx, ['Month', 'Plan']].reset_index(drop=True)

# Define the order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'Month' column to categorical data type with specified order
highest_avg_plan['Month'] = pd.Categorical(highest_avg_plan['Month'], categories=month_order, ordered=True)

# Sort the DataFrame based on the categorical 'Month' column
df_sorted = highest_avg_plan.sort_values(by='Month').reset_index(drop=True)

df_sorted.head(12)
```

```python
# Define the order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'Month' column to categorical data type with specified order
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
```

```python
df = df.sort_values(by=["Plan", "YEAR", "Month"]).reset_index(drop=True)
```

```python
def compound_per_plan(df):
    unique_plans = df['Plan'].unique()
    compound_per_plan = {}

    for plan in unique_plans:
        plan_df = df[df['Plan'] == plan]
        initial_amount = 1000  # Initial investment
        compound_interest = initial_amount

        for i in range(len(plan_df)):
            compound_interest *= (1 + plan_df['Value'].iloc[i])

        compound_per_plan[plan] = compound_interest

    return compound_per_plan
```

```python
compound_values = compound_per_plan(df)
for plan, compound_value in compound_values.items():
    print(f"Compound interest for {plan}: {compound_value}")
```
